```python
#standard imports

import pandas as pd
import numpy as np
```

```python
#import data from excel


# setup filename and object
fileName = "BrazilCOVIDData.xlsx"
xls = pd.ExcelFile(fileName)

# import individual sheets
covid_data = pd.read_excel(xls, 'Brazil Covid-19 data')
state_temp = pd.read_excel(xls, 'Temperature by State')
state_stats = pd.read_excel(xls, 'Brazil State Stats')
city_area = pd.read_excel(xls, 'City area')
```

```python
#merge state temperature and city area data on city name
temp_area_df= pd.merge(state_temp, city_area, how='outer', left_on="CITY", right_on="City")

#print dataframe
temp_area_df
```

```python
#list of values from 'IS_CAPITOL' column if city is a capital city
is_capital = ['Y', 'y']

#filter out data for capital cities
temp_area_capital_df = temp_area_df[temp_area_df['IS_CAPITOL'].isin(is_capital)]

#filter out certain columns from dataframe 
temp_area_capital_df = temp_area_capital_df[['STATE_ABBR','STATE', 'CITY', 'IS_CAPITOL','ANNUAL', 'SQ_KM']]

#print dataframe
temp_area_capital_df
```

```python
#create list of capital cities
capital_cities= ['Rio Branco', 'Maceió', 'Macapá', 'Manaus', 'Salvador', 'Fortaleza', 'Brasília', 'Vitória', 'Goiânia',
                'São Luís', 'Cuiabá', 'Campo Grande', 'Belo Horizonte', 'Belém', 'João Pessoa', 'Curitiba', 'Recife',
                 'Teresina', 'Rio de Janeiro', 'Natal', 'Porto Alegre', 'Porto Velho', 'Boa Vista', 'Florianópolis',
                 'São Paulo', 'Aracaju', 'Palmas']

#filter covid data for municipalities that match capital cities
covid_data_df = covid_data[covid_data['Municipality'].isin(capital_cities)]

#print dataframe
covid_data_df
```

```python
#merge covid dataframe with the temperature and area data frame

#merge the two dataframes to ensure city and states match
covid_area_temp_df = pd.merge(temp_area_capital_df, covid_data_df, how='outer', left_on=["CITY", 'STATE_ABBR'], right_on=["Municipality", 'State'])

#filter out capital cities only from the dataframe
covid_area_temp_df = covid_area_temp_df[covid_area_temp_df['IS_CAPITOL'].isin(is_capital)]

#print dataframe
covid_area_temp_df
```

```python
#create days column by iterating range of values through column

#import itertools
from itertools import cycle

#create range
c = cycle(range(0,151))

#create and iterate through days column
covid_area_temp_df['days'] = [next(c) for _ in range(len(covid_area_temp_df))]

#print dataframe
covid_area_temp_df
```

```python
#rename two columns
covid_area_temp_df.rename(columns={'ANNUAL': 'temp', 'Population as of 2019': 'pop'},inplace = True)

#create new columns

#create days_sq column by squaring days column
covid_area_temp_df['days_sq'] = covid_area_temp_df['days']**2

#create days_cube column by cubing days column
covid_area_temp_df['days_cube'] = covid_area_temp_df['days'] **3

#creating pop_dense column by dividing population by area; also converting columns to floats for calculations
covid_area_temp_df['pop_dense'] = (covid_area_temp_df['pop'].astype(float)) / (covid_area_temp_df['SQ_KM'].astype(float))

#creating pop_dense_sq column by squaring the population density
covid_area_temp_df['pop_dense_sq'] = covid_area_temp_df['pop_dense']**2

#print dataframe
covid_area_temp_df
```

```python
# Get Final Features DataFrame

features = covid_area_temp_df[['days_cube', 'days_sq', 'days', 'temp', 'pop_dense_sq', 'pop_dense', 'pop']]

# Get Final Response DataFrame

response = covid_area_temp_df[['Accumulated cases']]
```
